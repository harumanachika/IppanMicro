```python
#from google.colab import drive
#drive.mount('/content/drive')

import pandas as pd
import pandas_profiling as pdp
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ippan_shugyou/ippan_2012shugyou_dataset.csv", encoding="shift-jis")
df.tail()
```

```python
df = df.drop(df.columns[[0,1]], axis = 1).dropna()
df.tail()
#pdp.ProfileReport(df)
```

```python
%matplotlib inline
plt.scatter(df[["NDI"]], df[["C"]])
plt.title("Scatter plot of C vs NDI")
plt.xlabel("NDI")
plt.ylabel("C")
```

```python
df[["NDI", "C"]].corr()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

X = df[["NDI"]].values
Y = df[["C"]].values

lr.fit(X, Y)

print(lr.coef_)
print(lr.intercept_)
```
