```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import urllib
import json

from pandas_estat import set_appid
from pandas_estat import read_statslist
from pandas_estat import read_statsdata
```

```python
set_appid(appId)
lfs_list = pd.DataFrame(read_statslist(lfs))
display(lfs_list.columns)
lfs_list = lfs_list[['TABLE_INF','TITLE','CYCLE']].loc[lfs_list['CYCLE']=='月次']
lfs_list.to_csv("output/lfs_list.csv", encoding = 'utf-8')
```

```python
pp = read_statsdata(lfs_list['TABLE_INF'][327]) 
pp['Date'] = pd.to_datetime(pp['時間軸（月次）'], format='%Y年%m月')
pp.set_index('Date', inplace=True)
pp = pp.loc[pp['性別']=='総数']
pp = pp[['就業状態','value']]
pp = pp.pivot(columns='就業状態', values='value')
display(pp)
pp.to_csv("output/pp_data.csv")
```
